```python
import pandas as pd
```

```python
titanic = pd.read_csv('../data/aulas/titanic.csv', delimiter= ";")
```

```python
titanic.head()
```

```python
titanic.sex # categoria de feminino e masculino é dificil de ser interpretada pelo computador
```

```python
pd.get_dummies(titanic.sex).astype(int)
```

```python
sex = pd.get_dummies(titanic.sex).astype(int) # tirara categoria de sexo pois não nos ajuda mais
```

```python
titanic[['female', 'male']] = sex
```

```python
titanic.head()
```

```python
titanic.drop(['sex'], axis=1, inplace=True)  # tirara categoria male
```

```python
titanic.drop(['male'], axis=1, inplace=True)
```

```python
titanic.head()
```

```python
titanic.loc[(titanic.female == 0) & (titanic.who == 'woman') | (titanic.female == 1) & (titanic.who == 'man')] # colunas redundantes/ retire
```

```python
titanic.loc[(titanic.survived == 0) & (titanic.alive == 'yes') | (titanic.survived == 1) & (titanic.alive == 'no')] # colunas redundantes/ retire
```

```python
titanic.drop(['who', 'alive'], axis=1, inplace=True) #retirando as colunas who e alive
```

```python
titanic
```

```python
embarque_num = pd.get_dummies(titanic.embark_town).astype(int) # cada passageiro embarcou em uma cidade, se não foi em nenhuma das duas foi na outra
```

```python
titanic[['Queenstown', 'Southampton']] = embarque_num .drop(['Cherbourg'], axis=1)
```

```python
titanic.drop(['embark_town'], axis=1, inplace=True)
```
