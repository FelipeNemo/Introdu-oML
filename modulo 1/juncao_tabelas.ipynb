```python
import pandas as pd
```

```python
voos = pd.read_csv('../data/aulas/juncao_tabelas/voos.csv', delimiter= ";")
voos.head()
```

```python
qnt_voos = pd.read_csv('../data/aulas/juncao_tabelas/quatidade_de_voos.csv', delimiter= ";")
qnt_voos.tail()
```

```python
pd.merge(voos, qnt_voos, on='month') # merge em month
```

```python
produtos1 = pd.read_csv('../data/aulas/juncao_tabelas/produtos1.csv', delimiter= ";", index_col=['id']) # O id serÃ¡ o index da tabela(linhas)
produtos1.head()
```

```python
produtos2 = pd.read_csv('../data/aulas/juncao_tabelas/produtos2.csv', delimiter= ";", index_col=['id'])
produtos2.head()
```

```python
produtos = pd.concat([produtos1, produtos2]) # colocou uma embaixo da outra
produtos.head()
```

```python
vendas = pd.read_csv('../data/aulas/juncao_tabelas/vendas.csv', delimiter= ";")
vendas.head()
```

```python
vendas = vendas.join(produtos, on='produto_id') # o produto_id e o nome do produto na mesma coluna, o preÃ§o do sabonete equivale ao preÃ§o do sabonete nÃ£o a quantidade total
```

```python
vendas.head()
```

```python
vendas['total'] = vendas.quantidade * vendas.preco # operaÃ§Ãµes com colunas inteiras
```

```python
vendas.head()
```

```python
print(type(vendas['total']))
```

```python
# Garantir que as colunas quantidade e preco sejam numÃ©ricas
vendas['quantidade'] = pd.to_numeric(vendas['quantidade'], errors='coerce')
vendas['preco'] = pd.to_numeric(vendas['preco'], errors='coerce')

# Criar a coluna total multiplicando quantidade por preco
vendas['total'] = vendas['quantidade'] * vendas['preco']

# Exibir o resultado
print(vendas)
```
